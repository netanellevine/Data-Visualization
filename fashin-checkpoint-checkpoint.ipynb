## loading the data
We load the data through "tensorflow" and keep our final test aside untill the end so we could check our model better.

```python
(x_train_full, y_train_full), (x_test_final, y_test_final) = fashion_mnist.load_data()
x_train_full.shape
```

The data shape is 28*28 for all the images so we dont need to resize or change them.

```python
x_train_full = pd.DataFrame(x_train_full.reshape(60000, 28*28))
x_test_final = pd.DataFrame(x_test_final.reshape(10000, 28*28))
```

### now we will work only on the train data and at the end we check our accuracy on the final test

```python
x_train_full
```

We have 784 features which is a lot for data, later we will decrease the column number with PCA.

#### Before we start cleaning the data let's take a look over few images

```python
plt.figure(figsize=(9,9))
images_data=x_train_full[100:200].values
imgs=[image.reshape(28,28) for image in images_data]
imgs.append(np.zeros((28, 28 * 0)))
row_images = []
for row in range(10):
    rimages = imgs[row * 10 : (row + 1) * 10]
    row_images.append(np.concatenate(rimages, axis=1))
image = np.concatenate(row_images, axis=0)
plt.imshow(image, cmap = mpl.cm.binary)
plt.axis("off")
plt.show()
```

```python
missing_vals = pd.DataFrame(x_train_full.isnull().sum(), columns = ['Amount of missing values:'])
missing_vals['Percentage of missing values:']= (missing_vals / len(x_train_full)).round(4) * 100
print(x_train_full.isnull().sum().sum())
missing_vals
```

#### great we can see that our data is already clean and there is no missing values 

### Spliting the data into train and test
We chose to split 10% of our data to testing becuse we have 60,000 images, so 6,000 it is enough for testing.

```python
x_train, x_test, y_train, y_test = train_test_split(x_train_full, y_train_full, test_size=0.1, random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# PCA and scaling the data
### We now reduce features from our data and scale our data so that each features has the same effect on our model.


```python
from sklearn.pipeline import Pipeline

scaling_pipeline = Pipeline(
    [
        ('std_scaler', StandardScaler()),
        ('pca',PCA(n_components=0.90, random_state=0,svd_solver='auto')),       
    ]
)

x_train_scaled = scaling_pipeline.fit_transform(x_train)
x_test_scaled = scaling_pipeline.transform(x_test)
```

```python
x_train_scaled.shape
```

As we can see our number of features drops from 784 to 137, this is a significant difference that makes our models work faster 
and yet enough features so that our results do not hurt too much.

## Models
### now let's start training models!
#### first we create a function that let us evaluate each model easily 

```python
def results(model):
    y_pred=model.predict(x_test_scaled)
    plt.figure(figsize=(5,4))
    conf =confusion_matrix(y_test, y_pred)
    print("Confusion Matrix: ")
    sns.heatmap(conf, cmap = "Blues", annot=True, fmt = '.0f')
    plt.xlabel('True label')
    plt.ylabel('Predicted label')
    plt.show()
    print(classification_report(y_test, y_pred,digits=4))
```

### K Nearest Neighbors - KNN


```python
KNN =  KNeighborsClassifier(n_neighbors=53,weights='distance',n_jobs=-1)

KNN.fit(x_train_scaled,y_train)
KNN.score(x_test_scaled,y_test)
```

```python
scores = []
for i in range(45,55):
    knn_i = KNeighborsClassifier(n_neighbors = i)
    knn_i.fit(x_train_scaled,y_train)
    scores.append(knn_i.score(x_test_scaled,y_test))
    
plt.plot(range(45,55),scores)
plt.xlabel("Range")
plt.ylabel("Score")
plt.show()
```

### Logistic Regression



We chose solver=saga because it work faster with large datasets

```python
modelLogReg  = LogisticRegression(random_state=42,solver='saga',n_jobs=-1)
modelLogReg .fit(x_train_scaled, y_train)
modelLogReg .score(x_test_scaled, y_test)
```

### Gaussian Naive Bayes


```python
from sklearn.naive_bayes import GaussianNB
gaussian_nb =  GaussianNB()

gaussian_nb.fit(x_train_scaled,y_train)
gaussian_nb.score(x_test_scaled,y_test)
```

### Random Forest Classifier

First we start with the basic model

```python
Random_Forest=RandomForestClassifier()
Random_Forest.fit(x_train_scaled, y_train)
y_pred=Random_Forest.predict(x_test_scaled)
metrics.accuracy_score(y_test, y_pred)
```

#### Fine tuning the random forest model using random GridSearch.
We use the random grid search to get the best parameters for the model

```python
Random_Forest_params = {
     'n_estimators': [150,250,400],
     'max_depth':[None,30,50],
     'max_features': ['auto','sqrt'],
     'random_state':[42]
}

Random_Forest_grid_search = RandomizedSearchCV(
    RandomForestClassifier(n_jobs=-1),
    Random_Forest_params,
    cv=4,
    scoring='accuracy',
    return_train_score=True,
    random_state=42
)

Random_Forest_grid_search.fit(x_train_scaled, y_train)
print(Random_Forest_grid_search.best_estimator_)
print(Random_Forest_grid_search.best_score_)
```

Now we will choose the best parameters that we found and create another Random Forest Classifier.

```python
Best_Random_Forest = RandomForestClassifier(max_features='sqrt', n_estimators=400,
                       n_jobs=-1, random_state=42)
Best_Random_Forest.fit(x_train_scaled, y_train)
y_pred_forest = Best_Random_Forest.predict(x_test_scaled)

results(Best_Random_Forest)
```

### AdaBoost


```python
ada_clf = AdaBoostClassifier(
 DecisionTreeClassifier(max_depth=3), n_estimators=150,
 algorithm="SAMME.R", learning_rate=0.5)

ada_clf.fit(x_train_scaled, y_train)
prediction_test = ada_clf.predict(x_test_scaled)

results(ada_clf)
```

### XGBoost

First we start with the basic model

```python
xg_boost = XGBClassifier()
xg_boost.fit(x_train_scaled,y_train)
xg_boost.score(x_test_scaled, y_test)
```

#### Fine tuning the XGBoost model using random GridSearch.
We use the random grid search to get the best parameters for the model

```python
xgb_params = {
        'max_depth': [3,6,10],
        'learning_rate': [0.01, 0.05, 0.1],
        'n_estimators': [100, 500, 1000],
        'colsample_bytree': [0.3, 0.7],
    }

XGBGridsearch = RandomizedSearchCV(XGBClassifier(n_jobs=-1), xgb_params,scoring='accuracy',cv=3, random_state=0,n_jobs=-1)
XGBGridsearch.fit(x_train_scaled, y_train)

XGBGridsearch.best_estimator_
print(XGBGridsearch.best_estimator_)
print(XGBGridsearch.best_score_)
```

Now we will choose the best parameters that we found and create another XGBoost.

```python
best_xgb = xgboost.XGBClassifier(base_score=0.5, booster='gbtree', callbacks=None,
              colsample_bylevel=1, colsample_bynode=1, colsample_bytree=0.7,
              early_stopping_rounds=None, enable_categorical=False,
              eval_metric=None, gamma=0, gpu_id=-1, grow_policy='depthwise',
              importance_type=None, interaction_constraints='',
              learning_rate=0.1, max_bin=256, max_cat_to_onehot=4,
              max_delta_step=0, max_depth=10, max_leaves=0, min_child_weight=1,
              missing=nan, monotone_constraints='()', n_estimators=1000,
              n_jobs=-1, num_parallel_tree=1, predictor='auto', random_state=0,
              reg_alpha=0, reg_lambda=1)
best_xgb.fit(x_train_scaled, y_train)
y_pred = best_xgb.predict(x_test_scaled)

results(best_xgb)
```

###  Voting

#### soft classifier

```python
voting_clf = VotingClassifier(
estimators=[('rf', Best_Random_Forest),('knn', KNN),('xgb_bag',
BaggingClassifier(best_xgb))],voting='soft',weights=[1,4,1],n_jobs=-1)
voting_clf.fit(x_train_scaled, y_train)
results(voting_clf)
```

#### hard classifier

```python
voting_clf = VotingClassifier(
estimators=[('rf', Best_Random_Forest),('knn', KNN),('xgb_bag',
BaggingClassifier(best_xgb))],voting='hard',weights=[1,4,1],n_jobs=-1)

voting_clf.fit(x_train_scaled, y_train)
results(voting_clf)
```

###  Stacking


```python
estimators = [
    ('knn' , KNN),
    ('rf', Best_Random_Forest),
    ('xgb' , best_xgb)
]
stack = StackingClassifier(
    estimators=estimators, final_estimator=xg_boost
)
stack.fit(x_train_scaled,y_train)
results(stack)
```

### Bagging


```python
bag_clf = BaggingClassifier(
 xg_boost, n_estimators=30,
 random_state=40, bootstrap_features=True, n_jobs=-1)
bag_clf.fit(x_train_scaled, y_train)
results(bag_clf)
```
